```python
conversions_by_contacts = df.groupby(
    by='campaign'
)['conversion'].sum() / df.groupby(
    by='campaign'
)['conversion'].count() * 100
pd.DataFrame(conversions_by_contacts)
```

```python
ax = df.groupby('conversion').count()['contact'].plot(
    kind='line',
    grid=True,
    figsize=(10,7),
    title='conversion by campaign'
)

ax.set_xlabel('conversion')
ax.set_ylabel('contact')
plt.show()
```

# pivot table

why we not use the count because the count cannot take a NA so we use len

```python
status_df=pd.pivot_table(df, values='y',index='marital',columns='conversion',aggfunc=len)
status_df
```

```python
status_df.columns=['conversion','non_conversion']
```

```python
status_df
```

```python
status_df.plot(kind='pie',
      figsize=(15,7),
      startangle=90,
      subplots=True,
      autopct=lambda x: '%0.1f%%'%x
)
plt.show
```
